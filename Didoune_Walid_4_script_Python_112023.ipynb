```python
url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {
    "ingr" : "champagne"
}

headers = {
    "x-rapidapi-key" : "bd005cbe18msh77d510ee310220cp1d7198jsnccb86496515c",
    "x-rapidapi-host" : "edamam-food-and-grocery-database.p.rapidapi.com" 
}
```

```python
response = requests.get(url, headers=headers, params=querystring)
data = response.json()
data
```

```python
filtered_data = []

if "hints" in data:
    for item in data["hints"]:
        champagne_data = item.get("food", {})
        filtered_info = {
            "foodId" : champagne_data.get("foodId"),
            "label" : champagne_data.get("label"),
            "category" : champagne_data.get("category"),
            "foodContentsLabel" : champagne_data.get("foodContentsLabel"),
            "image" : champagne_data.get("image")
        }
        filtered_data.append(filtered_info)

filtered_data
```

```python
with open("data/champagne_rapidapi_data.csv", "w", newline="") as csv_file:
    column_names = [
        "foodId",
        "label",
        "category",
        "foodContentsLabel", 
        "image"
    ]
    writer = csv.DictWriter(csv_file, fieldnames=column_names)
    writer.writeheader()

    for row in filtered_data[:10]:
        writer.writerow(row)

champagne_data = pd.read_csv("data/champagne_rapidapi_data.csv")
champagne_data
```
